```python
import pandas as pd
import numpy as np
data=pd.read_csv("netflix_data.csv")
```

```python
data["cast"]=data["cast"].replace(np.nan, "None")
```

```python
data.head(5)
```

```python
def num_cast(var1):
    if var1=="None":
        return "0"
    else:
        return len(var1.split(", "))
```

```python
data["Num of Cast"]=data["cast"].apply(num_cast)
```
